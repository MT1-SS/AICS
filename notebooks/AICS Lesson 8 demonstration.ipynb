```python
# AI for Incident Response - Class 8 Demonstration
# Course: AI in Cybersecurity
# Topic: Accelerating Security Operations with Intelligent Automation

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import random
from sklearn.ensemble import RandomForestClassifier, IsolationForest
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import warnings
warnings.filterwarnings('ignore')

# Set style for better visualizations
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

print("=== AI FOR INCIDENT RESPONSE DEMONSTRATION ===")
print("Class 8: Accelerating Security Operations with Intelligent Automation\n")

# ============================================================================
# SECTION 1: GENERATING SYNTHETIC SECURITY EVENT DATA
# ============================================================================

print("1. GENERATING SYNTHETIC SECURITY EVENT DATA")
print("=" * 50)

def generate_security_events(n_events=1000):
    """Generate synthetic security events for demonstration"""
    
    # Event types and their characteristics
    event_types = {
        'Malware Detection': {'severity': 'High', 'auto_response': True, 'false_positive_rate': 0.05},
        'Phishing Email': {'severity': 'Medium', 'auto_response': True, 'false_positive_rate': 0.10},
        'Suspicious Login': {'severity': 'Medium', 'auto_response': False, 'false_positive_rate': 0.15},
        'Network Anomaly': {'severity': 'Low', 'auto_response': False, 'false_positive_rate': 0.20},
        'DDoS Attack': {'severity': 'Critical', 'auto_response': True, 'false_positive_rate': 0.02},
        'Data Exfiltration': {'severity': 'Critical', 'auto_response': True, 'false_positive_rate': 0.03},
        'Privilege Escalation': {'severity': 'High', 'auto_response': False, 'false_positive_rate': 0.08}
    }
    
    events = []
    start_time = datetime.now() - timedelta(days=30)
    
    for i in range(n_events):
        event_type = random.choices(
            list(event_types.keys()),
            weights=[0.25, 0.20, 0.30, 0.15, 0.05, 0.03, 0.02]
        )[0]
        
        characteristics = event_types[event_type]
        
        # Generate event timestamp
        event_time = start_time + timedelta(
            minutes=random.randint(0, 30*24*60)
        )
        
        # Determine if this is a true positive or false positive
        is_true_positive = random.random() > characteristics['false_positive_rate']
        
        event = {
            'event_id': f"EVT_{i+1:06d}",
            'timestamp': event_time,
            'event_type': event_type,
            'severity': characteristics['severity'],
            'source_ip': f"192.168.{random.randint(1,254)}.{random.randint(1,254)}",
            'destination_ip': f"10.0.{random.randint(1,254)}.{random.randint(1,254)}",
            'user_id': f"user_{random.randint(1,500):03d}",
            'asset_id': f"asset_{random.randint(1,100):03d}",
            'confidence_score': random.uniform(0.6, 0.95) if is_true_positive else random.uniform(0.3, 0.7),
            'auto_response_enabled': characteristics['auto_response'],
            'is_true_positive': is_true_positive,
            'response_time_minutes': None,
            'status': 'Open'
        }
        
        events.append(event)
    
    return pd.DataFrame(events)

# Generate sample data
security_events = generate_security_events(1000)
print(f"Generated {len(security_events)} security events")
print("\nSample events:")
print(security_events.head())

print(f"\nEvent type distribution:")
print(security_events['event_type'].value_counts())

# ============================================================================
# SECTION 2: AI-POWERED INCIDENT TRIAGE AND PRIORITIZATION
# ============================================================================

print("\n\n2. AI-POWERED INCIDENT TRIAGE AND PRIORITIZATION")
print("=" * 50)

def ai_incident_triage(events_df):
    """AI-powered incident triage and prioritization"""
    
    # Create features for ML model
    le_event_type = LabelEncoder()
    le_severity = LabelEncoder()
    
    events_df['event_type_encoded'] = le_event_type.fit_transform(events_df['event_type'])
    events_df['severity_encoded'] = le_severity.fit_transform(events_df['severity'])
    
    # Calculate priority score using multiple factors
    severity_weights = {'Critical': 4, 'High': 3, 'Medium': 2, 'Low': 1}
    events_df['severity_weight'] = events_df['severity'].map(severity_weights)
    
    # Priority score calculation
    events_df['priority_score'] = (
        events_df['confidence_score'] * 0.4 +
        events_df['severity_weight'] / 4 * 0.6
    )
    
    # Assign priority levels
    events_df['priority'] = pd.cut(
        events_df['priority_score'],
        bins=[0, 0.3, 0.6, 0.8, 1.0],
        labels=['P4-Low', 'P3-Medium', 'P2-High', 'P1-Critical']
    )
    
    return events_df, le_event_type, le_severity

# Apply AI triage
security_events, event_encoder, severity_encoder = ai_incident_triage(security_events)

print("Priority distribution after AI triage:")
print(security_events['priority'].value_counts())

# Visualize priority distribution
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
priority_counts = security_events['priority'].value_counts()
plt.pie(priority_counts.values, labels=priority_counts.index, autopct='%1.1f%%')
plt.title('Incident Priority Distribution')

plt.subplot(1, 2, 2)
sns.scatterplot(data=security_events, x='confidence_score', y='priority_score', 
                hue='severity', style='event_type', alpha=0.7)
plt.title('Confidence vs Priority Score')
plt.xlabel('Confidence Score')
plt.ylabel('Priority Score')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()

# ============================================================================
# SECTION 3: AUTOMATED INCIDENT RESPONSE WORKFLOWS (PLAYBOOKS)
# ============================================================================

print("\n\n3. AUTOMATED INCIDENT RESPONSE WORKFLOWS")
print("=" * 50)

class IncidentResponsePlaybook:
    """Automated incident response playbook system"""
    
    def __init__(self):
        self.playbooks = {
            'Malware Detection': self._malware_playbook,
            'Phishing Email': self._phishing_playbook,
            'DDoS Attack': self._ddos_playbook,
            'Data Exfiltration': self._data_exfiltration_playbook,
            'Suspicious Login': self._suspicious_login_playbook
        }
        self.actions_log = []
    
    def execute_playbook(self, incident):
        """Execute appropriate playbook for incident"""
        event_type = incident['event_type']
        
        if event_type in self.playbooks and incident['auto_response_enabled']:
            actions = self.playbooks[event_type](incident)
            
            # Log actions
            for action in actions:
                self.actions_log.append({
                    'incident_id': incident['event_id'],
                    'timestamp': datetime.now(),
                    'action': action,
                    'status': 'Executed'
                })
            
            return actions
        else:
            return ['Manual review required']
    
    def _malware_playbook(self, incident):
        """Malware detection playbook"""
        return [
            f"Isolate affected asset: {incident['asset_id']}",
            f"Block source IP: {incident['source_ip']}",
            f"Quarantine malicious files",
            f"Run full system scan on {incident['asset_id']}",
            f"Update threat intelligence feeds",
            f"Notify security team"
        ]
    
    def _phishing_playbook(self, incident):
        """Phishing email playbook"""
        return [
            f"Block sender email domain",
            f"Remove email from all mailboxes",
            f"Reset credentials for affected user: {incident['user_id']}",
            f"Add indicators to threat intelligence",
            f"Send security awareness notification"
        ]
    
    def _ddos_playbook(self, incident):
        """DDoS attack playbook"""
        return [
            f"Enable DDoS protection",
            f"Block attacking IPs: {incident['source_ip']}",
            f"Scale up infrastructure",
            f"Notify ISP and CDN providers",
            f"Monitor traffic patterns"
        ]
    
    def _data_exfiltration_playbook(self, incident):
        """Data exfiltration playbook"""
        return [
            f"Immediately block data transfer",
            f"Isolate affected systems: {incident['asset_id']}",
            f"Preserve forensic evidence",
            f"Disable user account: {incident['user_id']}",
            f"Contact legal team",
            f"Notify data protection officer"
        ]
    
    def _suspicious_login_playbook(self, incident):
        """Suspicious login playbook"""
        return [
            f"Require MFA for user: {incident['user_id']}",
            f"Monitor user activity closely",
            f"Check for privilege escalation attempts",
            f"Review access logs",
            f"Consider temporary account suspension"
        ]

# Create playbook system and execute responses
playbook_system = IncidentResponsePlaybook()

# Execute playbooks for high-priority incidents
high_priority_incidents = security_events[
    security_events['priority'].isin(['P1-Critical', 'P2-High'])
].head(10)

print("Executing automated response playbooks for high-priority incidents:")
print("-" * 60)

for idx, incident in high_priority_incidents.iterrows():
    actions = playbook_system.execute_playbook(incident)
    print(f"\nIncident: {incident['event_id']} - {incident['event_type']}")
    print(f"Priority: {incident['priority']}")
    print("Automated Actions:")
    for i, action in enumerate(actions, 1):
        print(f"  {i}. {action}")

# ============================================================================
# SECTION 4: AI-DRIVEN THREAT INTELLIGENCE CORRELATION
# ============================================================================

print("\n\n4. AI-DRIVEN THREAT INTELLIGENCE CORRELATION")
print("=" * 50)

def generate_threat_intelligence():
    """Generate synthetic threat intelligence data"""
    
    threat_indicators = {
        'IP': [f"192.168.{random.randint(1,254)}.{random.randint(1,254)}" for _ in range(100)],
        'Domain': [f"malicious-domain-{i}.com" for i in range(50)],
        'Hash': [f"{''.join(random.choices('abcdef0123456789', k=32))}" for _ in range(75)],
        'Email': [f"phishing{i}@suspicious-domain.com" for i in range(30)]
    }
    
    ti_data = []
    for indicator_type, indicators in threat_indicators.items():
        for indicator in indicators:
            ti_data.append({
                'indicator': indicator,
                'type': indicator_type,
                'threat_type': random.choice(['Malware', 'Phishing', 'Botnet', 'APT', 'Ransomware']),
                'confidence': random.uniform(0.6, 0.95),
                'first_seen': datetime.now() - timedelta(days=random.randint(1, 365)),
                'source': random.choice(['Commercial Feed', 'OSINT', 'Internal', 'Government'])
            })
    
    return pd.DataFrame(ti_data)

def correlate_events_with_threat_intel(events_df, ti_df):
    """Correlate security events with threat intelligence"""
    
    # Create correlation based on IP addresses
    correlations = []
    
    for idx, event in events_df.iterrows():
        # Check source IP against threat intel
        source_matches = ti_df[
            (ti_df['indicator'] == event['source_ip']) & 
            (ti_df['type'] == 'IP')
        ]
        
        # Check destination IP against threat intel
        dest_matches = ti_df[
            (ti_df['indicator'] == event['destination_ip']) & 
            (ti_df['type'] == 'IP')
        ]
        
        matches = pd.concat([source_matches, dest_matches])
        
        if not matches.empty:
            for _, match in matches.iterrows():
                correlations.append({
                    'event_id': event['event_id'],
                    'indicator': match['indicator'],
                    'indicator_type': match['type'],
                    'threat_type': match['threat_type'],
                    'ti_confidence': match['confidence'],
                    'correlation_score': (event['confidence_score'] + match['confidence']) / 2
                })
    
    return pd.DataFrame(correlations)

# Generate threat intelligence and correlate
threat_intel = generate_threat_intelligence()
correlations = correlate_events_with_threat_intel(security_events, threat_intel)

print(f"Generated {len(threat_intel)} threat intelligence indicators")
print(f"Found {len(correlations)} correlations between events and threat intel")

if len(correlations) > 0:
    print("\nTop threat intelligence correlations:")
    top_correlations = correlations.nlargest(5, 'correlation_score')
    print(top_correlations[['event_id', 'indicator', 'threat_type', 'correlation_score']])

# Visualize threat intelligence
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
ti_type_counts = threat_intel['type'].value_counts()
plt.bar(ti_type_counts.index, ti_type_counts.values)
plt.title('Threat Intelligence by Indicator Type')
plt.xticks(rotation=45)

plt.subplot(2, 2, 2)
threat_type_counts = threat_intel['threat_type'].value_counts()
plt.pie(threat_type_counts.values, labels=threat_type_counts.index, autopct='%1.1f%%')
plt.title('Threat Types Distribution')

plt.subplot(2, 2, 3)
if len(correlations) > 0:
    correlation_scores = correlations['correlation_score']
    plt.hist(correlation_scores, bins=20, alpha=0.7)
    plt.title('Correlation Score Distribution')
    plt.xlabel('Correlation Score')
    plt.ylabel('Frequency')

plt.subplot(2, 2, 4)
source_counts = threat_intel['source'].value_counts()
plt.bar(source_counts.index, source_counts.values)
plt.title('Threat Intel Sources')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# ============================================================================
# SECTION 5: ANOMALY DETECTION FOR SOC OPERATIONS
# ============================================================================

print("\n\n5. ANOMALY DETECTION FOR SOC OPERATIONS")
print("=" * 50)

def detect_anomalies_in_events(events_df):
    """Use machine learning to detect anomalous security events"""
    
    # Prepare features for anomaly detection
    features = events_df[['event_type_encoded', 'severity_encoded', 'confidence_score']].copy()
    
    # Add time-based features
    events_df['hour'] = pd.to_datetime(events_df['timestamp']).dt.hour
    events_df['day_of_week'] = pd.to_datetime(events_df['timestamp']).dt.dayofweek
    
    features['hour'] = events_df['hour']
    features['day_of_week'] = events_df['day_of_week']
    
    # Use Isolation Forest for anomaly detection
    iso_forest = IsolationForest(contamination=0.1, random_state=42)
    anomaly_labels = iso_forest.fit_predict(features)
    
    events_df['is_anomaly'] = anomaly_labels == -1
    
    return events_df

# Detect anomalies
security_events = detect_anomalies_in_events(security_events)

anomaly_count = security_events['is_anomaly'].sum()
print(f"Detected {anomaly_count} anomalous events out of {len(security_events)} total events")

# Show some anomalous events
if anomaly_count > 0:
    print("\nSample anomalous events:")
    anomalous_events = security_events[security_events['is_anomaly']].head()
    print(anomalous_events[['event_id', 'event_type', 'severity', 'confidence_score', 'hour']])

# ============================================================================
# SECTION 6: SOC DASHBOARD SIMULATION
# ============================================================================

print("\n\n6. SOC DASHBOARD SIMULATION")
print("=" * 50)

def create_soc_dashboard_data(events_df):
    """Create data for SOC dashboard visualization"""
    
    # Calculate key metrics
    total_events = len(events_df)
    high_priority = len(events_df[events_df['priority'].isin(['P1-Critical', 'P2-High'])])
    auto_resolved = len(events_df[events_df['auto_response_enabled']])
    anomalies = events_df['is_anomaly'].sum()
    
    # Mean Time To Response (MTTR) simulation
    events_df['mttr_hours'] = np.where(
        events_df['auto_response_enabled'],
        np.random.uniform(0.1, 0.5, len(events_df)),  # Automated: 6-30 minutes
        np.random.uniform(1, 8, len(events_df))       # Manual: 1-8 hours
    )
    
    avg_mttr = events_df['mttr_hours'].mean()
    
    return {
        'total_events': total_events,
        'high_priority_events': high_priority,
        'auto_resolved_events': auto_resolved,
        'anomalous_events': anomalies,
        'avg_mttr_hours': avg_mttr,
        'automation_rate': (auto_resolved / total_events) * 100
    }

# Generate SOC metrics
soc_metrics = create_soc_dashboard_data(security_events)

print("SOC DASHBOARD METRICS:")
print("-" * 30)
print(f"Total Security Events: {soc_metrics['total_events']:,}")
print(f"High Priority Events: {soc_metrics['high_priority_events']:,}")
print(f"Auto-Resolved Events: {soc_metrics['auto_resolved_events']:,}")
print(f"Anomalous Events: {soc_metrics['anomalous_events']:,}")
print(f"Average MTTR: {soc_metrics['avg_mttr_hours']:.2f} hours")
print(f"Automation Rate: {soc_metrics['automation_rate']:.1f}%")

# Create comprehensive SOC dashboard visualization
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Security Operations Center (SOC) Dashboard', fontsize=16, fontweight='bold')

# 1. Event Volume Over Time
events_by_hour = security_events.groupby(
    pd.to_datetime(security_events['timestamp']).dt.floor('H')
).size()

axes[0, 0].plot(events_by_hour.index, events_by_hour.values, linewidth=2)
axes[0, 0].set_title('Event Volume Over Time')
axes[0, 0].set_xlabel('Time')
axes[0, 0].set_ylabel('Number of Events')
axes[0, 0].tick_params(axis='x', rotation=45)

# 2. Priority Distribution
priority_data = security_events['priority'].value_counts()
axes[0, 1].pie(priority_data.values, labels=priority_data.index, autopct='%1.1f%%')
axes[0, 1].set_title('Incident Priority Distribution')

# 3. MTTR Comparison (Automated vs Manual)
mttr_comparison = security_events.groupby('auto_response_enabled')['mttr_hours'].mean()
labels = ['Manual Response', 'Automated Response']
axes[0, 2].bar(labels, [mttr_comparison[False], mttr_comparison[True]], 
               color=['red', 'green'], alpha=0.7)
axes[0, 2].set_title('Mean Time To Response (MTTR)')
axes[0, 2].set_ylabel('Hours')

# 4. Event Types Distribution
event_types = security_events['event_type'].value_counts()
axes[1, 0].barh(event_types.index, event_types.values)
axes[1, 0].set_title('Security Event Types')
axes[1, 0].set_xlabel('Count')

# 5. Confidence Score vs Priority
scatter_data = security_events.copy()
scatter_data['priority_numeric'] = scatter_data['priority'].map({
    'P4-Low': 1, 'P3-Medium': 2, 'P2-High': 3, 'P1-Critical': 4
})

scatter = axes[1, 1].scatter(scatter_data['confidence_score'], 
                            scatter_data['priority_numeric'],
                            c=scatter_data['is_anomaly'].astype(int),
                            cmap='coolwarm', alpha=0.6)
axes[1, 1].set_title('Confidence Score vs Priority')
axes[1, 1].set_xlabel('Confidence Score')
axes[1, 1].set_ylabel('Priority Level')
axes[1, 1].set_yticks([1, 2, 3, 4])
axes[1, 1].set_yticklabels(['P4-Low', 'P3-Medium', 'P2-High', 'P1-Critical'])

# 6. Key Performance Indicators
kpi_labels = ['Total\nEvents', 'High Priority\nEvents', 'Automated\nResponses', 'Anomalies\nDetected']
kpi_values = [soc_metrics['total_events'], soc_metrics['high_priority_events'], 
              soc_metrics['auto_resolved_events'], soc_metrics['anomalous_events']]

bars = axes[1, 2].bar(kpi_labels, kpi_values, color=['blue', 'orange', 'green', 'red'], alpha=0.7)
axes[1, 2].set_title('Key Performance Indicators')
axes[1, 2].set_ylabel('Count')

# Add value labels on bars
for bar, value in zip(bars, kpi_values):
    height = bar.get_height()
    axes[1, 2].text(bar.get_x() + bar.get_width()/2., height + height*0.01,
                   f'{value}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# ============================================================================
# SECTION 7: MACHINE LEARNING MODEL FOR INCIDENT CLASSIFICATION
# ============================================================================

print("\n\n7. MACHINE LEARNING MODEL FOR INCIDENT CLASSIFICATION")
print("=" * 50)

def train_incident_classifier(events_df):
    """Train ML model to classify incident severity"""
    
    # Prepare features
    features = ['event_type_encoded', 'confidence_score', 'hour', 'day_of_week']
    X = events_df[features]
    y = events_df['severity']
    
    # Split data
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42, stratify=y
    )
    
    # Train Random Forest classifier
    clf = RandomForestClassifier(n_estimators=100, random_state=42)
    clf.fit(X_train, y_train)
    
    # Make predictions
    y_pred = clf.predict(X_test)
    
    # Calculate accuracy
    accuracy = (y_pred == y_test).mean()
    
    print(f"Incident Classification Model Accuracy: {accuracy:.3f}")
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
    
    # Feature importance
    feature_importance = pd.DataFrame({
        'feature': features,
        'importance': clf.feature_importances_
    }).sort_values('importance', ascending=False)
    
    print("\nFeature Importance:")
    print(feature_importance)
    
    return clf, feature_importance

# Train the model
classifier, feature_importance = train_incident_classifier(security_events)

# Visualize feature importance
plt.figure(figsize=(10, 6))
plt.bar(feature_importance['feature'], feature_importance['importance'])
plt.title('Feature Importance for Incident Classification')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ============================================================================
# SECTION 8: AUTOMATED RESPONSE EFFECTIVENESS ANALYSIS
# ============================================================================

print("\n\n8. AUTOMATED RESPONSE EFFECTIVENESS ANALYSIS")
print("=" * 50)

def analyze_response_effectiveness(events_df):
    """Analyze the effectiveness of automated vs manual responses"""
    
    # Compare automated vs manual response metrics
    response_analysis = events_df.groupby('auto_response_enabled').agg({
        'mttr_hours': ['mean', 'std'],
        'is_true_positive': 'mean',
        'priority_score': 'mean'
    }).round(3)
    
    response_analysis.columns = ['_'.join(col).strip() for col in response_analysis.columns]
    
    print("Response Effectiveness Analysis:")
    print("-" * 40)
    print(response_analysis)
    
    # Calculate improvement metrics
    auto_mttr = events_df[events_df['auto_response_enabled']]['mttr_hours'].mean()
    manual_mttr = events_df[~events_df['auto_response_enabled']]['mttr_hours'].mean()
    mttr_improvement = ((manual_mttr - auto_mttr) / manual_mttr) * 100
    
    print(f"\nKey Improvements with AI Automation:")
    print(f"MTTR Reduction: {mttr_improvement:.1f}%")
    print(f"Automated Response Rate: {(events_df['auto_response_enabled'].mean() * 100):.1f}%")
    
    return response_analysis

# Analyze response effectiveness
response_effectiveness = analyze_response_effectiveness(security_events)

# Create before/after comparison visualization
plt.figure(figsize=(12, 8))

# Response time comparison
plt.subplot(2, 2, 1)
manual_times = security_events[~security_events['auto_response_enabled']]['mttr_hours']
auto_times = security_events[security_events['auto_response_enabled']]['mttr_hours']

plt.hist([manual_times, auto_times], bins=20, alpha=0.7, 
         label=['Manual Response', 'Automated Response'])
plt.title('Response Time Distribution')
plt.xlabel('Hours')
plt.ylabel('Frequency')
plt.legend()

# Response effectiveness by event type
plt.subplot(2, 2, 2)
effectiveness_by_type = security_events.groupby(['event_type', 'auto_response_enabled'])['mttr_hours'].mean().unstack()
effectiveness_by_type.plot(kind='bar', ax=plt.gca())
plt.title('MTTR by Event Type and Response Method')
plt.ylabel('Hours')
plt.xticks(rotation=45)
plt.legend(['Manual', 'Automated'])

# Automation rate by severity
plt.subplot(2, 2, 3)
automation_by_severity = security_events.groupby('severity')['auto_response_enabled'].mean() * 100
plt.bar(automation_by_severity.index, automation_by_severity.values)
plt.title('Automation Rate by Severity')
plt.ylabel('Automation Rate (%)')

# Cost-benefit analysis (simulated)
plt.subplot(2, 2, 4)
analyst_cost_per_hour = 75  # USD
auto_incidents = len(security_events[security_events['auto_response_enabled']])
manual_incidents = len(security_events[~security_events['auto_response_enabled']])

auto_cost = auto_incidents * security_events[security_events['auto_response_enabled']]['mttr_hours'].mean() * analyst_cost_per_hour
manual_cost = manual_incidents * security_events[~security_events['auto_response_enabled']]['mttr_hours'].mean() * analyst_cost_per_hour

costs = [manual_cost, auto_cost]
labels = ['Manual Response\nCost', 'Automated Response\nCost']
colors = ['red', 'green']

plt.bar(labels, costs, color=colors, alpha=0.7)
plt.title('Response Cost Comparison')
plt.ylabel('Cost (USD)')

# Add cost savings text
savings = manual_cost - auto_cost
plt.text(0.5, max(costs) * 0.8, f'Savings: ${savings:,.0f}', 
         ha='center', fontsize=12, fontweight='bold',
         bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.7))

plt.tight_layout()
plt.show()

# ============================================================================
# SUMMARY AND KEY TAKEAWAYS
# ============================================================================

print("\n\n" + "=" * 70)
print("SUMMARY AND KEY TAKEAWAYS")
print("=" * 70)

print("""
🎯 CLASS 8 DEMONSTRATION SUMMARY:

1. AUTOMATED INCIDENT RESPONSE WORKFLOWS
   ✓ Generated synthetic security events with realistic characteristics
   ✓ Implemented AI-powered triage and prioritization system
   ✓ Created automated response playbooks for different incident types
   ✓ Demonstrated significant MTTR reduction through automation

2. AI-DRIVEN THREAT INTELLIGENCE CORRELATION
   ✓ Simulated threat intelligence feeds from multiple sources
   ✓ Automated correlation between security events and threat indicators
   ✓ Enhanced incident context through intelligent data fusion

3. ATTACK CONTAINMENT AND REMEDIATION
   ✓ Automated containment actions based on incident type
   ✓ Systematic remediation procedures for different attack vectors
   ✓ Reduced human error through standardized responses

4. SOC TRANSFORMATION WITH AI
   ✓ Comprehensive SOC dashboard with real-time metrics
   ✓ Anomaly detection for identifying unusual security events
   ✓ Machine learning for incident classification and prioritization
   ✓ Performance analytics showing automation benefits

KEY BENEFITS DEMONSTRATED:
""")

# Calculate and display key metrics
total_incidents = len(security_events)
automated_incidents = len(security_events[security_events['auto_response_enabled']])
automation_rate = (automated_incidents / total_incidents) * 100

avg_manual_mttr = security_events[~security_events['auto_response_enabled']]['mttr_hours'].mean()
avg_auto_mttr = security_events[security_events['auto_response_enabled']]['mttr_hours'].mean()
mttr_improvement = ((avg_manual_mttr - avg_auto_mttr) / avg_manual_mttr) * 100

high_priority_incidents = len(security_events[security_events['priority'].isin(['P1-Critical', 'P2-High'])])
detected_anomalies = security_events['is_anomaly'].sum()

print(f"📊 Automation Rate: {automation_rate:.1f}%")
print(f"⚡ MTTR Improvement: {mttr_improvement:.1f}% faster response")
print(f"🎯 High Priority Incidents: {high_priority_incidents} identified and prioritized")
print(f"🔍 Anomalies Detected: {detected_anomalies} unusual events flagged")
print(f"🤖 Playbook Actions: {len(playbook_system.actions_log)} automated responses executed")

print(f"""
🚀 NEXT STEPS FOR IMPLEMENTATION:
1. Integrate with existing SIEM/SOAR platforms
2. Develop custom playbooks for organization-specific threats
3. Implement continuous learning and model improvement
4. Establish metrics and KPIs for measuring AI effectiveness
5. Train SOC analysts on AI-augmented workflows

💡 REMEMBER: AI enhances human analysts, it doesn't replace them!
   The goal is to free up analysts for complex, strategic tasks
   while AI handles routine, repetitive incident response actions.
""")

print("\n" + "=" * 70)
print("END OF AI FOR INCIDENT RESPONSE DEMONSTRATION")
print("=" * 70)
```
